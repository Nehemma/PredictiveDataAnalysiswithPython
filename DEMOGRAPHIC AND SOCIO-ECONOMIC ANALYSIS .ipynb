From the above plot, and the correlation table, it was evident that gender inequality affects the well-being index negatively.

```python
color = ['black']
Train_data1.plot.scatter(x='Well-Being Index',y='Old age dependency ratio (old age (65 and older) per 100 creatures (ages 15-64))',marker='o',c=color)
plt.title('Old Age Dependency vs Well being index',fontsize=20)
plt.ylabel('Old Age Dependency')
plt.xlabel('Well Being')
plt.show()
```

From the above graph, it is evident that there is a weak positive relationship between Old Age Dependency and well being.
